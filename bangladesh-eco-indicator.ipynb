```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import adfuller, grangercausalitytests
from statsmodels.tsa.api import VAR
from statsmodels.tsa.arima.model import ARIMA
import statsmodels.api as sm
from sklearn.preprocessing import StandardScaler
from scipy import stats

# Set style
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

# Load the data
# Make sure the path to your CSV file is correct
df = pd.read_csv('/kaggle/input/ecoinvestigation/BD_economic_indicators.csv')

# Clean column names by removing leading/trailing whitespace
df.columns = df.columns.str.strip()

# More robust data cleaning section
cleaning_map = {
    'GDP': ',',
    'GDP per capita': ',',
    'GDP growth': '%',
    'Inflation rate': '%',
    'Unemployed rate': '%',
    'Total Investment': '%',
    'Government debt': '%'
}

for column, char_to_remove in cleaning_map.items():
    if column in df.columns:
        if pd.api.types.is_string_dtype(df[column]):
            if char_to_remove == ',':
                df[column] = df[column].str.replace(',', '', regex=False)
            elif char_to_remove == '%':
                df[column] = df[column].str.rstrip('%')
        df[column] = pd.to_numeric(df[column], errors='coerce')

# Convert Year to datetime and set it as the index
df['Year'] = pd.to_datetime(df['Year'], format='%Y')
df.set_index('Year', inplace=True)

print("Dataset Overview after cleaning:")
print(df.info())
print("\nFirst few rows:")
print(df.head())


# 1. TIME SERIES ANALYSIS
fig, axes = plt.subplots(4, 2, figsize=(15, 12))
fig.suptitle('Bangladesh Economic Indicators Time Series (1980-2019)', fontsize=16)
plot_idx = 0
# Plot only columns that are not completely empty
for column in df.dropna(axis=1, how='all').select_dtypes(include=[np.number]).columns:
    if plot_idx < 8:
        row, col = divmod(plot_idx, 2)
        axes[row, col].plot(df.index, df[column], linewidth=2)
        axes[row, col].set_title(column)
        axes[row, col].grid(True, alpha=0.3)
        axes[row, col].tick_params(axis='x', rotation=45)
        plot_idx += 1
for i in range(plot_idx, 8):
    row, col = divmod(i, 2)
    fig.delaxes(axes[row, col])
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()


# 2. CORRELATION ANALYSIS
# --- FIX STARTS HERE ---
# First, select only numeric columns and drop columns that are COMPLETELY empty
df_with_data = df.select_dtypes(include=[np.number]).dropna(axis=1, how='all')

# Now, create the correlation dataframe by dropping any ROWS with remaining NaNs
corr_df = df_with_data.dropna(axis=0)
# --- FIX ENDS HERE ---


print(f"\nColumns available for correlation analysis: {corr_df.columns.tolist()}")
print(f"Time period for correlation: {corr_df.index.min().year} to {corr_df.index.max().year}")

# Correlation matrix
correlation_matrix = corr_df.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0,
            square=True, fmt='.2f', annot_kws={'size': 10})
plt.title('Correlation Matrix of Economic Indicators')
plt.tight_layout()
plt.show()

# Pairplot for visual correlation analysis
sns.pairplot(corr_df, diag_kind='kde')
plt.suptitle('Pairplot of Economic Indicators', y=1.02)
plt.show()


# 3. ECONOMETRIC MODELING

# A. Stationarity Tests (Augmented Dickey-Fuller Test)
print("\nStationarity Tests (ADF):")
print("Null Hypothesis: The series has a unit root (non-stationary)")
print("="*50)
for column in corr_df.columns:
    result = adfuller(corr_df[column]) # .dropna() is redundant here as corr_df has no NaNs
    print(f"{column}:")
    print(f"ADF Statistic: {result[0]:.4f}")
    print(f"p-value: {result[1]:.4f}")
    print("Stationary" if result[1] < 0.05 else "Non-stationary")
    print("-"*30)


# B. VAR Model for multiple time series
# --- FIX: Removed 'Unemployed rate' as it has no data ---
var_columns = [col for col in ['GDP growth', 'Inflation rate', 'Total Investment']
               if col in corr_df.columns]

if len(var_columns) >= 2:
    var_data = corr_df[var_columns]
    print(f"\nVAR analysis period: {var_data.index.min().year} to {var_data.index.max().year}")
    print(f"VAR variables: {var_columns}")
    
    var_data_stationary = var_data.copy()
    for col in var_data.columns:
        result = adfuller(var_data[col])
        if result[1] > 0.05:
            print(f"{col} is non-stationary, applying differencing")
            var_data_stationary[col] = var_data[col].diff()
    var_data_stationary = var_data_stationary.dropna()
    
    if len(var_data_stationary) > 5:
        model = VAR(var_data_stationary)
        results = model.fit(maxlags=2, ic='aic')
        print("\nVAR Model Summary:")
        print(results.summary())
        print("\nGranger Causality Tests:")
        maxlag = 2
        for col in var_data_stationary.columns:
            for other_col in var_data_stationary.columns:
                if col != other_col:
                    gc_result = grangercausalitytests(var_data_stationary[[col, other_col]], maxlag=maxlag, verbose=False)
                    p_values = [gc_result[i+1][0]['ssr_chi2test'][1] for i in range(maxlag)]
                    if min(p_values) < 0.05:
                        print(f"{other_col} Granger-causes {col} (min p-value: {min(p_values):.4f})")
    else:
        print("Insufficient data for VAR analysis after differencing")
else:
    print("Insufficient variables for VAR analysis")


# C. ARIMA Model for GDP Growth forecasting
gdp_growth = df['GDP growth'].dropna()
try:
    model_arima = ARIMA(gdp_growth, order=(1,0,1))
    results_arima = model_arima.fit()
    print("\nARIMA Model for GDP Growth:")
    print(results_arima.summary())
    
    forecast_steps = 5
    forecast = results_arima.get_forecast(steps=forecast_steps)
    forecast_index = pd.date_range(start=gdp_growth.index[-1], periods=forecast_steps+1, freq='AS-JAN')[1:]
    
    plt.figure(figsize=(12, 6))
    plt.plot(gdp_growth.index, gdp_growth, 'b-', label='Actual GDP Growth', linewidth=2)
    plt.plot(forecast_index, forecast.predicted_mean, 'r--', label='Forecast', linewidth=2)
    plt.fill_between(forecast_index, forecast.conf_int().iloc[:, 0], forecast.conf_int().iloc[:, 1], alpha=0.3, color='red')
    plt.title('GDP Growth Rate: Actual vs Forecast (ARIMA Model)')
    plt.xlabel('Year')
    plt.ylabel('GDP Growth Rate (%)')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()
except Exception as e:
    print(f"ARIMA modeling failed: {e}")


# D. Multiple Regression Analysis
reg_data = corr_df.copy()
if 'GDP growth' in reg_data.columns:
    # --- FIX: Removed 'Unemployed rate' as it has no data ---
    X_columns = [col for col in ['Inflation rate', 'Total Investment'] 
               if col in reg_data.columns and col != 'GDP growth']
    
    if X_columns:
        X = sm.add_constant(reg_data[X_columns])
        y = reg_data['GDP growth']
        model_ols = sm.OLS(y, X).fit()
        
        print("\nMultiple Regression Analysis:")
        print(f"Dependent variable: GDP Growth")
        print(f"Independent variables: {X_columns}")
        print("="*60)
        print(model_ols.summary())
        
        fig, axes = plt.subplots(2, 2, figsize=(12, 10))
        fig.suptitle('Regression Diagnostic Plots')
        sns.residplot(x=model_ols.fittedvalues, y=model_ols.resid, lowess=True, ax=axes[0,0], line_kws={'color': 'red', 'lw': 1})
        axes[0,0].set_title('Residuals vs Fitted')
        sm.qqplot(model_ols.resid, line='s', ax=axes[0,1])
        axes[0,1].set_title('Normal Q-Q Plot')
        sns.histplot(model_ols.resid, kde=True, ax=axes[1,0])
        axes[1,0].set_title('Residual Distribution')
        fig.delaxes(axes[1,1])
        plt.tight_layout(rect=[0, 0.03, 1, 0.95])
        plt.show()
    else:
        print("Insufficient independent variables for regression analysis")
else:
    print("GDP growth data not available for regression analysis")


# E. Economic Growth Decomposition Analysis
gdp_growth_series = df['GDP growth'].dropna()
if len(gdp_growth_series) > 10:
    try:
        decomposition = seasonal_decompose(gdp_growth_series, model='additive', period=5, extrapolate_trend='freq') 
        fig = decomposition.plot()
        fig.set_size_inches(12, 8)
        plt.suptitle('GDP Growth Rate Decomposition', y=1.02)
        plt.tight_layout(rect=[0, 0.03, 1, 1])
        plt.show()
    except ValueError as e:
        print(f"Seasonal decomposition not possible: {e}")

# Calculate basic statistics
print("\nDescriptive Statistics:")
print(corr_df.describe())

# Calculate growth rates and correlations
if 'GDP growth' in corr_df.columns:
    print(f"\nAverage GDP Growth: {corr_df['GDP growth'].mean():.2f}%")
if 'Inflation rate' in corr_df.columns:
    print(f"Average Inflation: {corr_df['Inflation rate'].mean():.2f}%")
if 'Total Investment' in corr_df.columns and 'GDP growth' in corr_df.columns:
    print(f"Correlation between GDP Growth and Investment: {corr_df['GDP growth'].corr(corr_df['Total Investment']):.3f}")

print("\nAnalysis completed successfully!")
```
